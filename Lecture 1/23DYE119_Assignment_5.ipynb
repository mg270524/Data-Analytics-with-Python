```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie','May', 'Bonn', 'Chan','Alex', 'Bobby', 'Cio','Leon', 'Peter', 'Ray'],
    'Age': [25, 30, 32, 24, 36, 32, 29, 31, 32, 27, 32, 34],
    'Salary': [50000, 68000, 75000, 55000, 60000, 170000, 50000, 160000, 170000,250000, 160000, 70000]
}
df = pd.DataFrame(data)
print(df)
```

```python
df['Bonus'] = df['Salary'] * 0.10  
df['Total Compensation'] = df['Salary'] + df['Bonus']
df
```

```python
df.sort_values(by='Salary', ascending=False)  # Sort by Salary descending
```

```python
df.isnull().sum()  # Count missing values
df.fillna(0, inplace=True)  # Replace NaNs with 0
df.dropna(inplace=True)  # Drop rows with missing values
df
```

```python
import pandas as pd

#  Grouping
data = {
    'Department': ['IT', 'HR', 'IT', 'HR', 'Finance'],
    'Salary': [60000, 50000, 70000, 55000, 80000],
    'Bonus': [5000, 4000, 6000, 4500, 7000]
}
df = pd.DataFrame(data)

grouped_df = df.groupby('Department').mean()
print(grouped_df)
```

```python

df.groupby('Department').agg({'Salary': 'mean', 'Bonus': 'sum'})  # Multiple aggregations
```

```python
df.groupby('Department').count()  # Count of employees in each department
```

```python
df.groupby('Department')['Salary'].sum()  # Total salary per department
```

```python
# Merging
df1 = pd.DataFrame({
    'Employee': ['Alice', 'Bob', 'Charlie'],
    'Department': ['IT', 'HR', 'Finance']
})

df2 = pd.DataFrame({
    'Employee': ['Alice', 'Bob', 'Charlie'],
    'Salary': [60000, 50000, 90000]
})

merged_df = pd.merge(df1, df2, on='Employee')
print(merged_df)
```

```python
# Treating missing value wuth different method.
import numpy as np
import pandas as pd
np.random.seed(420)
data = {
    'Employee_ID': np.arange(1, 101),
    'Department': np.random.choice(['IT', 'HR', 'Finance'], size=100),
    'Salary': np.random.randint(40000, 100000, size=100).astype(float),
    'Bonus': np.random.randint(2000, 10000, size=100).astype(float)
}
data['Salary'][3] = np.nan  
data['Salary'][4] = np.nan 
data['Bonus'][7] = np.nan  
data['Bonus'][17] = np.nan  
 

df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df.isnull().any(axis=1)
```

```python
df1 = df.fillna(0) # fill with 0
df1
```

```python
df.ffill(inplace=True)  # Forward fill (use previous row's value)
df
```

```python
df.bfill(inplace=True)  # Backward fill (use next row's value)
df
```
