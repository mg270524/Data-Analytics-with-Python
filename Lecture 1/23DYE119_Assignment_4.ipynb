```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.normal(10,5,100)
y = 3 + .5*x + np.random.normal(0,1,100)
vec = np.array([x,y])
plt.scatter(x,y,c='b',s=5)
plt.show()
plt.hist(x, bins=20, color='blue', edgecolor='black', alpha=0.7)
plt.title('Histogram of x')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
x_sorted = np.sort(x)
cdf = np.arange(1, len(x_sorted) + 1) / len(x_sorted)
plt.plot(x_sorted, cdf, label="CDF of x", color="blue")
plt.title("Cumulative Distribution Function (CDF) of x")
plt.xlabel("Value")
plt.ylabel("CDF")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def pi(a):
    inx = []
    iny = []
    outx = []
    outy = []
    inside = 0
    for a in range(a):
        x= np.random.uniform(-1, 1)
        y= np.random.uniform(-1, 1)
        if x**2 + y**2 <= 1:
            inside += 1
            inx.append(x)
            iny.append(y)
        else:
            outx.append(x)
            outy.append(y)
    
    pi= (inside / a) * 4
    return pi, inx, iny, outx, outy

a = int(input("Enter the number of points: "))
pi, inx, iny, outx, outy = pi(a)

plt.figure(figsize=(6,6))
plt.scatter(inx, iny, c='b', s=5, label='Inside Circle')
plt.scatter(outx, outy, c='r', s=5, label='Outside Circle')
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title(f"Monte Carlo Approximation of π with {a} Points")
plt.show()
```

## Page rank algorithm.

```python
import sympy as sp
import random
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from sympy import Matrix, zeros
from random import sample, random
```

### For Adjecency Matrix

```python

n, m = 20, 20
total= 150
A= zeros(n, m)
positions = sample(range(n * m), total)
for pos in positions:
    i, j = divmod(pos, m)
    A[i, j] = 1
G = nx.Graph()
for i in range(n):
    for j in range(m):
        if A[i, j] != 0:
            G.add_edge(i, j) 
plt.figure(figsize=(8, 8))
nx.draw(G, 
        with_labels=False,
        node_size=50, 
        node_color='blue', 
        edge_color='red')
plt.show()
```

```python
B= sp.Matrix([[2000],[3000],[1500],[4000],[1000],[500],[250],[1500],[900],[800],[350],[2400],[4250],[5025],[1525],[2900],[2450],[540],[780],[3800]])
```

```python
B # Initial Rank Vector
```

### Transition Matrix

```python
import random
import sympy as sp

for i in range(A.rows):
    for j in range(A.cols):
        if A[i, j] == 1:
            A[i, j] = random.randint(0, 1000)

for j in range(A.cols):
    column_sum = sum(A[i, j] for i in range(A.rows))
    if column_sum != 0:
        for i in range(A.rows):
            A[i, j] = A[i, j] / column_sum
```

```python
A # Transition Matrix
```

```python
s = np.ones((20,20))/20
S = sp.Matrix(s)
```

```python
S #Teleportation Matrix
```

```python
d = 0.85 # Damping Factor
G = d * A + (1 - d) * S
```

```python
G # Google Matrix
```
