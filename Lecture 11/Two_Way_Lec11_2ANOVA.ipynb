**Notations:**

* $\overline{y}_{i.}$ is the mean of $i$-th treatment. That is, 
$\overline{y}_{i.}=\frac{1}{c}\sum_{j=1}^c y_{ij}$.
*  $\overline{y}_{.j}$ is the mean of $j$-th block. That is, 
* $\overline{y}_{.j}=\frac{1}{r}\sum_{i=1}^r y_{ij}$.

*  $\overline{y}_{..}$ is the grand mean. That is,  $\overline{y}_{..}=\frac{1}{rc}\sum_{i=1}^r\sum_{j=1}^c y_{ij}$.

*  $SST= \sum_{i=1}^r\sum_{j=1}^c \left(y_{ij}-\overline{y}_{..}\right)^2$ 
*   $SSR= \sum_{i=1}^r\sum_{j=1}^c \left(\overline{y}_{i.}-\overline{y}_{..}\right)^2=c\sum_{i=1}^r\left(y_{i.}-\overline{y}_{..}\right)^2$ 

*  $SSC= \sum_{i=1}^r\sum_{j=1}^c \left(\overline{y}_{.j}-\overline{y}_{..}\right)^2=r\sum_{j=1}^c\left(y_{.j}-\overline{y}_{..}\right)^2$ 
*  $SSE= \sum_{i=1}^r\sum_{j=1}^c \left(y_{ij}-\overline{y}_{i.}-\overline{y}_{.j}+\overline{y}_{..}\right)^2$ 

* $SSE=SST-SSR-SSC$

*  $ MST:=\frac{SST}{rc-1}$
* ${MSR:=\frac{SSR}{r-1}}$

* ${MSC:=\frac{SSC}{c-1}}$

* ${MSE:=\frac{SSE}{(r-1)(c-1)}}$




**2-Way ANOVA Table**

\begin{array}{ccccc}
	Source~of 	& Sum~of & Degree~of & Mean~square &	 Ratio \\
	variation 	& square & freedom   &  						&		F			\\\hline\\
	Betwen~ Columns & SSC& c-1 & MSC=\frac{SSC}{c-1} & \frac{MSC}{MSE} \\[2mm]
	Betwen~ Rows  & SSR& r-1 & MSR=\frac{SSR}{r-1} & \frac{MSR}{MSE}\\[2mm]
	Residual & SSE& (r-1)(c-1) & MSE=\frac{SSE}{(r-1)(c-1)} & \\[2mm]
	Total& SST& rc-1 & MST=\frac{SST}{rc-1} & \\
\end{array}


**Shortcut Formulae**

$$
y_{..}=\sum \sum y_{ij,} \quad y_{i.}=\sum_{j=1}^c y_{ij}, \quad y_{.j}=\sum_{i=1}^r y_{ij}, \quad C=\frac{T^2}{rc}
$$ 

$$
SST=\sum\sum y_{ij}^2 -C, 
\quad 
SSC=\frac{\sum_{j=1}^{c}(T_{.j})^2}{r} -C 
$$

$$
SSR=\frac{\sum_{i=1}^{r}(T_{i.})^2}{c} -C, 
\quad 
SSE=SST-SSC-SSR
$$



**2-Way ANOVA Table for interaction and Replicates**

If there are two main factors $A$ and $B$ having $a$ and $b$ in numbers and there are $r$ replicates.  Let $y_{ijk}$ represent the observation in $i$-th item of factor $A$ and $j$-th item of factor $B$ and in $k$- replicate.


**2-Way ANOVA Table with Interaction**
\begin{array}{ccccc}
	Source~of 	& Sum~of & Degree~of & Mean square &	 Ratio \\[-1mm]
	variation 	& square & freedom   &  						&		F			\\\hline\\
	A & SSA& a-1 & MSA=\frac{SSA}{a-1} & \frac{MSA}{MSE} \\[4mm]
	B	  & SSB& b-1 & MSB=\frac{SSB}{b-1} & \frac{MSB}{MSE}\\[4mm]
	A\times B& SS(AB)& (a-1)(b-1) & MS(AB)=\frac{SS(AB)}{(a-1)(b-1)} & \frac{MS(AB)}{MSE} \\[4mm]
	Error  & SSE & ab(r-1) & MSE=\frac{SSE}{ab(r-1)} & \\[4mm]
	Total& SST& rab-1 & MST=\frac{SST}{rab-1} & \\
\end{array}


# Two-Way ANOVA using Python

### Example 1:
This is an experiment involving a storage
battery used in the launching mechanism of a shoulderfired
ground-to-air missile. Three material types can be
used to make the battery plates. The objective is to design a
battery that is relatively unaffected by the ambient temperature.
The output response from the battery is effective life
in hours. Three temperature levels are selected, and a factorial
experiment with four replicates is run. The data are as
follows:


```python
import numpy as np
import pandas as pd
material = ['M1']*12+['M2']*12+['M3']*12
temp = ['low']*2+['medium']*2+['high']*2
temp = temp*6
life = [130, 155, 34, 40, 20, 70,
  74, 180, 80, 75, 82, 58,
  150, 188, 136, 122, 25, 70,
  159, 126, 106, 115, 58, 45,
  138, 110, 174, 120, 96, 104,
  168, 160, 150, 139, 82, 60]
df = {'Material':material,'Temp':temp,'Life':life}
df = pd.DataFrame(df)
```

```python
df.columns=['Material','Temp','Life'] 
```

```python
df
```

```python
df.boxplot(column='Life', by='Material');
```

```python
df.boxplot(column='Life', by='Temp');
```

```python
from statsmodels.graphics.factorplots import interaction_plot
from matplotlib import pyplot as plt
```

```python
fig = interaction_plot(x=df['Material'], trace=df['Temp'], 
                       response=df['Life'], 
    colors=['red','green','blue'])
plt.show()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
model = ols('life ~ C(Material) + C(Temp)', 
            data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
model = ols('life ~ C(Material) + C(Temp) + C(Material):C(Temp)', 
            data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

## Step by Step Calculations

```python
import numpy as np
import pandas as pd
from scipy import stats
material = ['M1']*12+['M2']*12+['M3']*12
temp = ['low']*2+['medium']*2+['high']*2
temp = temp*6
replicates = ['1','2']*3+['3','4']*3
replicates = replicates*3
life = [130, 155, 34, 40, 20, 70,
  74, 180, 80, 75, 82, 58,
  150, 188, 136, 122, 25, 70,
  159, 126, 106, 115, 58, 45,
  138, 110, 174, 120, 96, 104,
  168, 160, 150, 139, 82, 60]
df = {'Material':material,'Temp':temp,'Life':life,'Replicates':replicates}
df = pd.DataFrame(df)
```

```python
N = len(df.Life)
p = len(df.Material.unique())
q = len(df.Temp.unique())
m = len(df.Replicates.unique())## No of replicateds
df_A = p - 1
df_B = q - 1
df_AB = df_A*df_B 
df_E = N - p*q
grand_mean = df['Life'].mean()
## Computation of SS_T
SS_T = sum((df.Life - grand_mean)**2)
## Computation of SS_A
SS_A = m*q*sum([(df[df.Material ==l].Life.mean()-grand_mean)**2 for l in df.Material.unique()])
## Computation of SS_B
SS_B = m*p*sum([(df[df.Temp ==t].Life.mean()-grand_mean)**2 for t in df.Temp.unique()])
## Computation of SS_E
ss = 0
for s in df.Material.unique():
    for t in df.Temp.unique():
        ss=ss+np.sum([((df[(df.Material==s) & (df.Temp==t)].Life)-df[(df.Material==s) & (df.Temp==t)].Life.mean())**2])
SS_E = ss
## Computation of SS_AB
SS_AB = SS_T-SS_A-SS_B-SS_E
MS_A = SS_A/df_A
MS_B = SS_B/df_B
MS_AB = SS_AB/df_AB
MS_E = SS_E/df_E
F_A = MS_A/MS_E
F_B = MS_B/MS_E
F_AB = MS_AB/MS_E
p_A = stats.f.sf(F_A, df_A, df_E)
p_B = stats.f.sf(F_B, df_B, df_E)
p_AB = stats.f.sf(F_AB, df_AB, df_E)
results = {'sum_sq':[SS_A, SS_B, SS_AB, SS_E],
           'df':[df_A, df_B, df_AB, df_E],
           'F':[F_A, F_B, F_AB, 'NaN'],
            'P(&gt;F)':[p_A, p_B, p_AB, 'NaN']}
columns=['sum_sq', 'df', 'F', 'P(&gt;F)']
aov_table1 = pd.DataFrame(results, columns=columns,
                          index=['Material', 'Temp', 
                          'Material:Tpmp', 'Residual'])
print(aov_table1)
```

### Example 2

```python
import pandas as pd
import seaborn as sns
```

```python
# load data file
d = pd.read_csv("https://reneshbedre.github.io/assets/posts/anova/twowayanova.txt", sep="\t")
# reshape the d dataframe suitable for statsmodels package 
# you do not need to reshape if your data is already in stacked format. Compare d and d_melt tables for detail 
# understanding 
d_melt = pd.melt(d, id_vars=['Genotype'], value_vars=['1_year', '2_year', '3_year'])
```

```python
d_melt
```

```python
d_melt.columns = ['Genotype', 'years', 'value']
```

```python
d_melt
```

```python
sns.boxplot(x="Genotype", y="value", hue="years", data=d_melt, palette="Set3") 
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
model = ols('value ~ C(Genotype) + C(years) + C(Genotype):C(years)', data=d_melt).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

```python
from statsmodels.graphics.factorplots import interaction_plot
import matplotlib.pyplot as plt
fig = interaction_plot(x=d_melt['Genotype'], trace=d_melt['years'], response=d_melt['value'], 
    colors=['#4c061d','#d17a22', '#b4c292'])
plt.show()
```
